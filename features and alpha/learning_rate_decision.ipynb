```python
import numpy as np
import matplotlib.pyplot as plt
from labfiles.lab_utils_multi import  load_house_data, run_gradient_descent
from labfiles.lab_utils_multi import  norm_plot, plt_equal_scale, plot_cost_i_w
from labfiles.lab_utils_common import dlc
np.set_printoptions(precision=2)
plt.style.use('labfiles/deeplearning.mplstyle')
```

```python
# load the dataset
X_train, y_train = load_house_data()
X_features = ['size(sqft)','bedrooms','floors','age']
```

```python
# plotting feature vs price 
fig,ax=plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X_train[:,i],y_train)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("Price (1000's)")
plt.show()
```

```python
# running gradient descent with alpha = 9.9e-7
#set alpha to 9.9e-7
_, _, hist = run_gradient_descent(X_train, y_train, 10, alpha = 9.9e-7)
```

```python
# It appears the learning rate is too high.
# The solution does not converge.
# Cost is increasing rather than decreasing
# plotiing learning curve and hhistory of gradient descent
plot_cost_i_w(X_train, y_train, hist)
```

```python
# trying smaller learning rate 
#set alpha to 9e-7
_,_,hist = run_gradient_descent(X_train, y_train, 40, alpha = 9e-7)
```

```python
# Cost is decreasing throughout the run showing that alpha is not too large. 
plot_cost_i_w(X_train, y_train, hist)
```

```python
# smaller value for alpha
#set alpha to 1e-7
_,_,hist = run_gradient_descent(X_train, y_train, 10, alpha = 1e-7)
```

```python
# visualizing the learning curve
plot_cost_i_w(X_train,y_train,hist)
```
