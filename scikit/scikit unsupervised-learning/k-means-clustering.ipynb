```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
# reading data from file
with open(r"data\Customers_data.csv") as f:
    reader = csv.reader(f)
    next(reader)
    
    data = []
    for row in reader:
        data.append([
            float(row[2]),
            float(row[4])
        ])
```

```python
# visualizing data
age = [row[0] for row in data]
score = [row[1] for row in data]

plt.title("Un-clustered data")
plt.xlabel('Age')  
plt.ylabel('Spending Score')
plt.scatter(age, score)
plt.show()
```

```python
# elbow method for finding k
css_list = []

for i in range(1, 11):
    k_means = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    k_means.fit(data)
    css_list.append(k_means.inertia_)
      
plt.title('Elbow Graph')  
plt.xlabel('No. of Clusters(k)')
plt.ylabel('Cluster Sum of Squares')
plt.plot(range(1, 11), css_list)
plt.show()
```

```python
# training K-Means model
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)  
predictions = kmeans.fit_predict(data)
```

```python
predictions
```

```python
list = []
for i in range(4):
    temp = []
    for j in range(len(predictions)):
        if predictions[j] == i:
            temp.append([age[j], score[j]])
    list.append(temp)

list0 = list[0]  
blue1, blue2 = [row[0] for row in list0],[row[1] for row in list0]

list1 = list[1] 
green1, green2 = [row[0] for row in list1],[row[1] for row in list1]

list2 = list[2] 
red1, red2 = [row[0] for row in list2],[row[1] for row in list2]

list3 = list[3] 
orange1, orange2 = [row[0] for row in list3],[row[1] for row in list3]
```

```python
# visualizing clusters
plt.title("Clusters")
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.scatter(blue1, blue2, s=100, c="blue", label = "1st cluster")
plt.scatter(green1, green2, s=100, c="green", label = "2nd cluster")
plt.scatter(red1, red2, s=100, c="red", label = "3rd cluster")
plt.scatter(orange1, orange2, s=100, c="orange", label = "4th cluster")
```
